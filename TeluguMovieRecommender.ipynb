```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

movies_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IndianMovies.csv')

movies_data.drop_duplicates()
movies_data.loc[0].dropna()
movies_data = movies_data.replace({'Rating(10)': {'-': '0'}}, regex=True)

movies_data["Rating(10)"] = pd.to_numeric(movies_data["Rating(10)"])
tmovies_data = movies_data.loc[movies_data['Language'].isin(['telugu'])]

selected_features = ['Movie Name','Genre','Votes','Rating(10)','Year']

tmovies_data = tmovies_data.replace({selected_features[0]: {'-': '0'}}, regex=True)
tmovies_data = tmovies_data.replace({selected_features[1]: {'-': '0'}}, regex=True)
tmovies_data = tmovies_data.replace({selected_features[2]: {'-': '0'}}, regex=True)
tmovies_data = tmovies_data.replace({selected_features[3]: {'-': '0'}}, regex=True)
tmovies_data = tmovies_data.replace({selected_features[4]: {'-': '0'}}, regex=True)

tmovies_data['ID'] = tmovies_data['ID'].str.replace(r'tt', '')

tmovies_data = tmovies_data.replace({'ID': {"-": '0'}}, regex=True)
tmovies_data['ID'] = pd.to_numeric(tmovies_data['ID'])

tmovies_data.reset_index(inplace = True)
tmovies_data.columns =['SNo',	'ID' , 'Movie Name',	'Year',	'Timing(min)',	'Rating(10)',	'Votes',	'Genre',	'Language']

# tmovies_data.head()

tmovies_genre = tmovies_data['Genre']
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(tmovies_genre)

similarity = cosine_similarity(feature_vectors)

movie_name = input(' Enter your favourite movie name : ')

list_of_all = tmovies_data['Movie Name'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all)

close_match = find_close_match[0]

index_of_movie = tmovies_data[tmovies_data['Movie Name'] == close_match].values[0]


similarity_score = list(enumerate(similarity[index_of_movie[0]]))

i = 1

sorted_sim_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
for movie in sorted_sim_movies:
  index = movie[0]
  title_from_index = tmovies_data[tmovies_data.index==index]['Movie Name'].values[0]
  if (i<20):
    print(i, '.',title_from_index)
    i+=1
```
